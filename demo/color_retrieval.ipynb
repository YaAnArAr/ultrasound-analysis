```python
import matplotlib.pyplot as plt

from src.videotools import read_video
from src.processing import get_frame_colors, PixelColor
```

```python
from IPython.display import clear_output 

PAUSE = 1/100


image, red, blue = [None] * 3
for frame in read_video('../data/image04.wmv'):
    clear_output(True)
    fig, ax = plt.subplots(ncols=3, figsize=(20, 40))
    colors = get_frame_colors(frame, sensivity=20)
    red_colors = colors == PixelColor.RED.value
    blue_colors = colors == PixelColor.BLUE.value
    image = ax[0].imshow(frame)
    red = ax[1].imshow(red_colors, cmap='gray')
    blue = ax[2].imshow(blue_colors, cmap='gray')
    plt.pause(PAUSE)
    plt.show()
```
