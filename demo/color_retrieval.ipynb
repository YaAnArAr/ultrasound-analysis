```python
import matplotlib.pyplot as plt
import numpy as np

from IPython.display import clear_output 

from src.videotools import read_video
from src.processing import get_frame_colors, PixelColor
```

```python
PAUSE = 1/100
SENSIVITY = 20


for frame in read_video('../data/image04.wmv'):
    clear_output(True)
    fig, ax = plt.subplots(ncols=2, figsize=(20, 40))
    colors = get_frame_colors(frame, sensivity=SENSIVITY)
    red_colors = colors == PixelColor.RED.value
    blue_colors = colors == PixelColor.BLUE.value
    mask = np.zeros_like(frame, dtype=np.uint8)
    mask[:, :, 0] += 255 * red_colors.astype('uint8')
    mask[:, :, -1] += 255 * blue_colors.astype('uint8')
    ax[0].imshow(frame)
    plt.axis('off')
    ax[1].imshow(mask)
    plt.axis('off')
    plt.pause(PAUSE)
    plt.show()
```
